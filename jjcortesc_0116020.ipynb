```python
'''
EJERCICIO # 02
Juan Jose Cortes
Leer dos números enteros y determinar si
la diferencia entre los dos es un número primo.

'''

# %% se definen las funciones


def ultimo_dig(x):
    ''' Funcion para encontrar el ultimo dígito '''
    ult = x % 10
    return ult

# %%###### ########## ######## PROGRAMA PRINCIPAL ######## ########## #########

while True:
    try:
        num1 = int(input("Dígite primer número entero: "))
        num2 = int(input("Dígite segundo numero entero: "))
        break
        # se define el rango del codigo
    except ValueError:
        print("No es un número entero, intente de nuevo ")
res = num1-num2    # Diferencia entre num1 y num2

if res < 0:        # El resultado se vuelve un valor positivo
    res = res*-1
n = 2              # El contador n empieza desde dos para
                   # poder evaluar los residuos
while res != n:
    if ultimo_dig(res) == 0:            # Se evalua el residuo para determinar
        print("\nEl número no es primo")  # si es primo, si es cero es primo,
        break                           # sino se aumenta n hasta que n sea
    n += 1                              # igual que la difirencia
else:
    print("\nEl número es primo")
```

```python
"""
EJERCICIO # 04
Juan Jose Cortes 

Haga un programa que lea las coordenadas (x1,y1,r1) y (x2,y2,r2)
que corresponden al centro y al radio de dos círculos.
Lea un punto de coordenadas (a,b).
Determina si (a,b) está contenido:
a) dentro del círculo 1;
b) dentro del círculo 2;
c) dentro de ambos círculos.
Recordemos que el círculo es el lugar geométrico de los puntos del plano
cuya distancia a otro punto fijo, llamado centro,
es menor o igual que una cantidad constante, llamada radio.

"""
# %% se importan las librerias
import math

# %% se definen las funciones
def dent_circ(a, b, x, y, r): 
    ''' Función para ver si un punto esta dentro de un círculo'''
    return x-r <= a <= x+r and y-r <= b <= y+r and math.hypot(a, b) <= r


# %%###### ########## ######## PROGRAMA PRINCIPAL ######## ########## #########

# Se ingresan  los círculos y el punto

while True:
    try:
        x1, y1, r1 = float(input("Dígite x1: ")),\
         float(input("Dígite y1: ")), float(input("Dígite r1: "))
        x2, y2, r2 = float(input("Dígite x2: ")),\
            float(input("Dígite y2: ")), float(input("Dígite r2: "))
        a, b = float(input("Dígite a: ")), float(input("Dígite b: "))
        break
    except ValueError:
        print("El número no es valido, intente de nuevo ")

# Se evalua en que situaciión se encuetra el punto
if dent_circ(a, b, x1, y1, r1):
    print("\nSe encuentra en el círculo de coordenadas x1,y1 y radio r1")
elif dent_circ(a, b, x2, y2, r2):
    print("\nSe encuentra en el círculo de coordenadas x2,y2 y radio r2 ")
elif dent_circ(a, b, x1, y1, r1) and dent_circ(a, b, x2, y2, r2):
    print("\nSe encuentra  en ambos circulos")
else:
    print("\nSe encuentra por fuera de los círculos")
```

```python
'''
EJERCICIO # 05
Juan Jose Cortes

Escriba un programa que reciba una cadena de texto y reporte:
a) cuántas letras vocales en mayúscula se entraron;
b) cuántas letras con tilde se entraron (minúsculas y mayúsculas);
c) cuántos dígitos se entraron,
d) cuántos espacios se entraron;
e) cuántas palabras reservadas se entraron.
Utilice para ello los métodos de cadenas especializados como .isupper(),
.isdigit(), etc. y la palabra reservada in.


'''
# %% se importan las librerias
import keyword

# %%###### ########## ######## PROGRAMA PRINCIPAL ######## ########## #########

# Inicialmente se pedirá al usuario ingresar la cadena de texto a evaluar.

texto = input('Ingrese un texto:')

# Se establecen los siguientes contadores, para cada tipo de variable.
palabra = ''  # Variable que extraerá palabra a palabra.
vocales_mayusculas = 0
letras_tilde = 0
digitos = 0
espacios = 0
palabras_reservadas = 0
# El siguiente ciclo leerá uno a uno los caracteres de la cadena y en caso de
# ser una uno de los caracteres buscados sumará a su respectivo contador.
for i in texto:
    if i.isdigit():  # Evaluá si el carácter extraido es un dígito.
        digitos += 1
    if i.isupper():  # Evaluá si el carácter extraido es un mayúscula.
        if ord(i) in (65, 69, 73, 79, 85):
            vocales_mayusculas += 1
    # Análogamente se evalúa para caracteres con tildes.
    if ord(i) in (225, 193, 201, 233, 205, 237, 211, 243, 218, 250):
        letras_tilde += 1
    # Ahora, se agruparán los caracteres formando palabras, para esto se ponen
    # las siguientes condiciones para que cada que aparezca un espacio o uno
    # de estos caracteres se deje agrupar la palabra y continue.
    if i != ' ' and i != '.' and i != ',' and i != '-' and i != '"':
        palabra = palabra + i
    if i == ' ' or i == '.' or i == ',' or i == '-' or i == '"':
        if i == ' ':
            espacios += 1
        # Luego, se revisa si la palabra es reservada.
        if palabra in (keyword.kwlist):
            palabras_reservadas += 1
        palabra = ''

# Finalmente, se imprime el resultado del analisis.
print(f'\nEn el texto hay {espacios} espacios, {digitos} dígitos ,' +
      f'{letras_tilde} letras con tilde, {palabras_reservadas} palabras' +
      f' reservadas y {vocales_mayusculas} vocales en mayúscula.')
```

```python
"""
EJERCICIO # 06
Juan Jose Cortes

Leer una cadena de texto y organizar alfabéticamente cada una de las
letras que la componen, repitiendo cada una tantas veces como se encuentra.
Por ejemplo, la cadena ‘tarea importante’ será ‘aaaeeimnoprrttt’.
(Note que no se incluyen los espacios).

"""
# %% se definen las funciones


def lista_cad(x):
    '''Funcion que convierte una lista en una cadena'''
    # se crea la lista vacia
    str1 = ""
    # se recorre la lista y se le agrega a la cadena obviando los espacios
    for ele in x:
        if ele == " ":
            continue
        else:
            str1 += ele
    # se retorna una cadena de texto
    return str1

# %%###### ########## ######## PROGRAMA PRINCIPAL ######## ########## #########


while True:
    try:
        cad = input("Dígite una cadena de texto: ")
        break
        # se define la cadena
    except ValueError:
        print("La cadena no es valida, intente de nuevo ")


lista = []
cad = cad.lower()  # Se convierte la cadena en minuscula
                   # para organizar las letras en orden alfabetico

for i in range(0, len(cad)):  # Se evalua cada letra de la cadena
    lista.append(cad[i])      # y se añade en una lista
orden = sorted(lista)         # se organiza la lista
final = lista_cad(orden)      # se convierte la lista en cadena

print(f"la cadena organizada alfabeticamente es:\n{final}")
```

```python
'''
EJERCICIO # 07
Juan Jose Cortes

Leer una lista de números ordenados de forma ascendente y verificar que está
este ordenada. Luego, leer un número e insértarlo en la lista en la posición
que le corresponde.


'''

# %% se definen las funciones


def lista_de_numeros(lista):
    '''Verifica que una lista tenga unicamente números.'''
    for i in lista:
        if type(i) != float:
            return False
    return True


def lista_ordenada(lista):
    '''Verifica que los elementos de una lista estén ordenados.'''
    return all([((lista[i]) < (lista[i+1])) for i in range(0, len(lista) - 1)])


def leer_numero(texto):
    '''Lee un número flotante.

       num = leer_numero()
    '''
    while True:  # Se verifica que el dato ingresado sea un número.
        try:
            num = float(input(texto))
            break
        except ValueError:
            print('Error: Ingrese un número.')
    return num


# %%###### ########## ######## PROGRAMA PRINCIPAL ######## ########## #########


# inicialmente se pide al usuario ingresar una lista.

while True:
    try:
        lista = eval(input("Ingrese una lista de números ordenados: "))
        if type(lista) != list:
            print('ERROR: No ha ingresado una lista.')
            continue
        for i in range(0, len(lista)):
            lista[i] = float(lista[i])
        if lista_de_numeros(lista):
            if lista_ordenada(lista):
                break
            else:
                print('Los elementos no están organizados ascendentemente.')
            continue
        else:
            print('ERROR: Ha ingresado elementos inválidos.')
    except KeyboardInterrupt:
        print('Acaba de presionar Ctrl+C')
    except SyntaxError:
        print('Esta ingresando elementos incorrectos.')
    except NameError:
        print('Esta ingresando elementos incorrectos.')
    except TypeError:
        print('Esta ingresando elementos incorrectos.')
    except ValueError:
        print('Esta ingresando elementos incorrectos.')


# %% se lee e ingresa un número en la posición correcta.
num = leer_numero('Ingrese un número: ')
for i in range(0, len(lista) - 1):
    if i == 0 and num < lista[i]:
        lista.insert(0, num)
    if lista[i] < num < lista[i + 1]:
        lista.insert((i + 1), num)
        break
    if i == (len(lista) - 2):
        lista.append(num)

# %% se imprime la lista definitiva.
print(f'\nLa lista queda de la siguiente manera:\n{lista}')
```

```python
"""
EJERCICIO # 13
Juan Jose Cortes

Diseñar una función que pida dos años y escriba cuántos años bisiestos
hay entre esas dos fechas (incluidos los dos años).

"""
# %%###### ########## ######## PROGRAMA PRINCIPAL ######## ########## #########

# se define la funcion


def cant_bis(año1, año2):
    '''Funcion para determinar cuantos años bisiestos
       hay entre dos años determinados'''

    def es_bis(x):  # se verifica que el año evaluado sea bisiesto

        '''  Funcíon para determinar si un año es bisiesto  '''
        return x % 400 == 0 or x % 100 != 0 and x % 4 == 0

    cant = 0  # Contador de años bisiestos

    for i in range(año1, año2+1):  # Se evalua cada año
        if es_bis(i):              # y se agrega al cotador de cantidad
            cant += 1               # si hay un año bisiesto
    return cant
```
